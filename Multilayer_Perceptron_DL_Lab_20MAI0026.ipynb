***4. Train Test Datasplit***



> 1.With 10% test data




```
# splitting into train and test datasets 90:10
X=mnist.drop(0,axis=1)
Y=mnist[0]
train_data, test_data, train_labels, test_labels = train_test_split(X, Y,test_size=0.1)
```



> 2.With 20% test data




```
# splitting into train and test datasets 80:20
X=mnist.drop(0,axis=1)
Y=mnist[0]
train_data, test_data, train_labels, test_labels = train_test_split(X, Y,
                            test_size=0.2)
```



> 3.With 30% test data




```
# splitting into train and test datasets 70:30
X=mnist.drop(0,axis=1)
Y=mnist[0]
train_data, test_data, train_labels, test_labels = train_test_split(X, Y,
                            test_size=0.3)
```



> 4.With 40% test data




```
# splitting into train and test datasets 60:40
X=mnist.drop(0,axis=1)
Y=mnist[0]
train_data, test_data, train_labels, test_labels = train_test_split(X, Y,
                            test_size=0.4)
```



> 5.With 50% test data




```
# splitting into train and test datasets 50:50
X=mnist.drop(0,axis=1)
Y=mnist[0]
train_data, test_data, train_labels, test_labels = train_test_split(X, Y,
                            test_size=0.5)
```

***5. Preprocessing the data***

```
# scaling the data
print("\n Roll No: 20MAI0026")
print("\n********************")
scaler = StandardScaler()
# we fit the train data
scaler.fit(train_data)
# scaling the train data
train_data = scaler.transform(train_data)
test_data = scaler.transform(test_data)
print(train_data[:3])
```

***6. Training the model:***

```
# Training the Model
print("\n Roll No: 20MAI0026")
print("\n********************")
# creating an classifier from the model:
mlp = MLPClassifier(hidden_layer_sizes=(100, ), max_iter=480, alpha=1e-4,
                    solver='sgd', verbose=10, 
                    tol=1e-4, random_state=1,
                    learning_rate_init=.1)

# let's fit the training data to our model
mlp.fit(train_data, train_labels)
```

***7. Testing the model:***

```
#Testing the model
predictions_train = mlp.predict(train_data)
predictions_test = mlp.predict(test_data)

```

***8. Evaluation of model metrics:***

```
#Evaluation of model metrics
print("\n Roll No: 20MAI0026")
print("\n********************")
#Confusion matrix and accuracy for the train model
print("Train Confusion Matrix: \n"+ str(confusion_matrix(predictions_train, train_labels))+"\n")
print("Training Accuracy: "+str(accuracy_score(predictions_train, train_labels)*100)+"%\n")
#Confusion matrix and accuracy for testing
print("Test Confusion Matrix: \n"+ str(confusion_matrix(predictions_test, test_labels)))
print("Testing Accuracy: "+str(accuracy_score(predictions_test, test_labels)*100)+"%")
```
